```python

import pandas as pd
import numpy as np
np.random.seed (1111)
import matplotlib.pyplot as plt
%matplotlib inline
import keras
from  keras.datasets import mnist
from  keras.models import Model
from  keras.layers import *
from  keras import optimizers
from  sklearn.model_selection import train_test_split
from  keras.utils.np_utils import to_categorical
```

```python
df_train = pd.read_csv ('/content/drive/MyDrive/Content/АВСОФТ_полная.csv') #импортим тренировочный набор
df_test = pd.read_csv ('/content/drive/MyDrive/Content/АВСОФТ_тест.csv')#импортим тестовый набор, который потом не использую) вместо этого я использовал train_test_split позже
```

```python
#Маловариативные данные перевожу в дамми код (Автора и Репозиторий)
#От даты (имхо) имеет смысл брать час загрузки или время от прошлой загруки. Я использую час загрузки
#Хеш код перевожу в десятичное число, хотя зависимости от хеша у количества багов не должно быть 

df_train = pd.get_dummies(df_train, columns=['repository_name'])
for i in range(len(df_train.commit_hash.values)):
  df_train.commit_hash.values[i] = int(df_train.commit_hash.values[i],16)/10000
for i in range(len(df_train.commit_date.values)):
  p1 = df_train.commit_date.values[i].find(' ')+1
  p2 = df_train.commit_date.values[i].find(':')
  df_train.commit_date.values[i] = int(df_train.commit_date.values[i][p1:p2],10)
df_train = pd.get_dummies(df_train, columns=['commit_date'])
df_train = pd.get_dummies(df_train, columns=['commit_author'])
df_train.commit_message = df_train.commit_message.str.len()
```

```python
#на всякий случай форматирую данные

df_features = df_train.iloc[:, 1:37]
df_label = df_train.iloc[:, 0]
X_test = df_test.iloc[:, 1:37]
X_train = df_train.iloc[:, 1:37] 
X_train = X_train.values.reshape(317, 36)
print(X_test.shape)
print(X_train.shape)
```

```python
#Отрезал Тестовый массив данных от Проверочного и рандомизировал это. (можно менять рандом стейт) 
X_train, X_cv, y_train, y_cv = train_test_split(df_features, df_label,
                                                test_size = 0.2,
                                                shuffle=True,
                                                random_state = 1111)
```

```python
#меняю формат данных на float32 делая все одноформатным

X_train = X_train.astype('float32')
X_cv = X_cv.astype('float32')
X_test = X_test.astype('float32')
```

```python
num_digits = 20
y_train = to_categorical(y_train, num_digits)
y_cv = to_categorical(y_cv, num_digits)
```

```python
n_input = 36
n_hidden_1 = 72  
n_hidden_2 = 72
num_digits = 20
```
